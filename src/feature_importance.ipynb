# Feature Importance assessment
## 1. Mutual Information Classifier

```python
def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_classif(X, y, discrete_features=discrete_features)
    # mi_scores = mutual_info_classif(X, y)
    mi_scores = pd.Series(mi_scores, name='MI Scores', index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = make_mi_scores(X, y, discrete_features)
mi_scores
    
```

```python
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    # print(f'width = {width}')
    # print(f'ticks = {ticks}')
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")


plt.figure(dpi=100, figsize=(8, 5))
plot_mi_scores(mi_scores)
```

## 2. Correlation Heatmap (Correlation Matrix)

```python
combined_df = pd.concat([X, y], axis=1)

corr_matrix = combined_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Matrix")
plt.show()
```

## 3. Permutation Importance

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance


X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=42)

model = RandomForestClassifier()
model.fit(X_train, y_train)

result = permutation_importance(model, X_test, y_test, n_repeats=20)
sorted_importances = result.importances_mean.argsort()

plt.barh(X_test.columns[sorted_importances], result.importances_mean[sorted_importances])
plt.title("Permutation Importance")
plt.show()
```

```python
import shap

explainer = shap.Explainer(model, X_train)
shap_values = explainer(X_train)

shap.summary_plot(shap_values, X_train)

shap.dependence_plot("FeatureName", shap_values, X_train)
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np


# # Get numeric values:
num_columns = [ col for col in X_train.columns if X_train[col].dtype in ['int', 'float']]
print(f'num_columns = {num_columns}')
# discrete_features = [ col for col in X.columns if X[col].dtype in ['int']]
# print(f'num_columns = {discrete_features}')
# # print(f'discrete_features = {discrete_features}')
scaler = StandardScaler()
X_train[num_columns] = scaler.fit_transform(X_train[num_columns])
# discrete_features


model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)

# Get the coefficients
feature_importance = np.abs(model_lr.coef_[0])

indices = np.argsort(feature_importance)[::-1]
feature_names = X_train.columns  # If using a DataFrame

plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.barh(range(X_train.shape[1]), feature_importance[indices], align="center")
plt.yticks(range(X_train.shape[1]), feature_names[indices])
plt.gca().invert_yaxis()  # Most important at the top
plt.show()
```
